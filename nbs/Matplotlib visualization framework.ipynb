```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
years = np.arange(2010, 2016)
years
```

```python
plt.figure(figsize=(10,10))

for cat_num in range(1,17):
    plt.subplot(4,4, cat_num)
    y_vals = np.arange(10,16) + (np.random.random(6) * cat_num / 4.)
    plt.plot(years, y_vals)
    plt.ylim(10,18)
    plt.xticks(years, [str(year) for year in years])
    plt.title(f'Category {cat_num}')
    
plt.tight_layout()
;
```
